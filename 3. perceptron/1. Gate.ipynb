### 2.3.1 간단한 구현부터
-----

```python
def AND1(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1 * w1 + x2 * w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
    
AND1(0,0) # 0
AND1(1,0) # 0
AND1(1,1) # 1
```

### 2.3.3 가중치와 편향 구현하기
------

```python
def AND(x1, x2):
    
    x = np.array([x1, x2])   # 파라메터
    w = np.array([0.5, 0.5]) # 가중치
    b = -0.7                 # 편향값
    
    tmp = np.sum(x*w)+b

    if tmp <= 0 :
        return 0
    elif tmp > 0 :
        return 1
    

AND(1,1)
```

```python
def NAND(x1, x2):
    
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(x*w) + b
    if tmp <= 0:
        return 0
    elif tmp > 0:
        return 1
    

NAND(1,1)
```

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(x*w)+b
    if tmp <= 0:
        return 0
    elif tmp > 0:
        return 1
    

OR(0,0)
```

```python
OR(1,0)
```

```python
OR(1,1)
```

```python
def NOR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.2
    tmp = np.sum(x*w)+b
    if tmp <= 0:
        return 0
    elif tmp > 0:
        return 1
    

NOR(0,0)
```

### 2.5.2 XOR 게이트 구현하기
-----


```python
def XOR(x1, x2):
    s1 = OR(x1, x2)
    s2 = NAND(x1, x2)
    y = AND(s1, s2)
    return y
```

```python
XOR(0, 1)
```
