```python
# csv파일 불러오기
df = pd.read_csv('news_data_ori_0701_0705.csv')



from konlpy.tag import Okt
from collections import Counter
from IPython.display import clear_output
import time


korean_stopwords_path = './korean_stopwords.txt'
with open(korean_stopwords_path, encoding='utf8') as f:
    stopwords = f.readlines()
stopwords = [x.strip() for x in stopwords]

nouns_cnt = 1
count_data = []


def get_nouns(data):
    global nouns_cnt, count_data
    
#     fchamgo = lambda x : x.split('/')
#     list(map(fchamgo,tmp_arr1, 1))   # map에 스칼라 숫자 전달
#     print(list(map(fchamgo,tmp_arr1)) )

    tmp_arr1 = data.split(' ')
    result_arr = []
    result_arr.extend([tmp1.split('/')[0] for tmp1 in tmp_arr1 if len(tmp1.split('/')) == 2 and 'NN' in tmp1.split('/')[1] ])
#     print([tmp1.split('/')[0] if len(tmp1.split('/')) == 2 and 'NN' in tmp1.split('/')[1] else 'd' for tmp1 in tmp_arr1 ])

    count_data.extend([x for x in result_arr])
#     count_data.extend([z for y in result_arr for z in y])

    clear_output(wait=True)
    per = round((nouns_cnt / df.shape[0])*100,3)
#     if per % 1 = 0 :
    print(f'형태소,불용어 제거중 .. {per} %')
    nouns_cnt+=1
    time.sleep(0.001)
    
    return result_arr

df.head()
```

```python
# df['nouns'] = df['ko_text'].apply(lambda x: get_nouns(x))
df['nouns'] = df['형태소'].apply(lambda x: get_nouns(x))
print(df.shape)
df.head()
```

## 3. 연관 키워드 추출하기

### [트위터 연관 키워드 분석]

```python
transactions = df['nouns'].tolist()
transactions = [transaction for transaction in transactions if transaction] # 공백 문자열을 방지합니다.and
# transactions
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori as apriori2
from mlxtend.frequent_patterns import association_rules

te = TransactionEncoder()
te_result = te.fit(transactions).transform(transactions)

print(te_result.shape[0]*te_result.shape[1])
```

```python
df2 = pd.DataFrame(te_result, columns=te.columns_)
print(df2)
print('======================')
df2['확진'].sum()
```
