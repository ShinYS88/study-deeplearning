```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터 준비
x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1 간격으로 생성
y = np.sin(x)

# 그래프 그리기
plt.plot(x,y)
plt.show()
```

```python
# cos 추가
y2 = np.cos(x)

# 그래프 그리기
plt.plot(x, y, label="sin")
plt.plot(x, y2, linestyle="--", label="cos") # cos 함수는 점선으로 그리기
plt.xlabel("x") # x축 이름
plt.ylabel("y") # y축 이름
plt.title("sin & cos") # 제목
plt.legend()
plt.show()
```
